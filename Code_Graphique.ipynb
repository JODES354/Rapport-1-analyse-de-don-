1) Vérification de la loi : FHWM $\propto E_{\gamma}^{-\frac{1}{2}}$

```python
print('Les barres d\'erreurs sont plus petites que la taille des points.')

# Régression linéaire
modeleReg_1=LinearRegression()

resulta_1 = modeleReg_1.fit(pd.DataFrame(np.log(E_4)), np.log(Incertitude_4/E_4))

# nuage de point
plt.scatter(np.log(E_4), np.log(Incertitude_4/E_4), color='r', marker='p', label=r'gain = 75', s=110)

# courbe de tendance
plt.plot(np.linspace(2.64, 4.1,  1000), modeleReg_1.predict(pd.DataFrame(np.linspace(2.64, 4.1, 1000))), color='black', linestyle='--', label='courbe de tendance')

# Titre et paramétres des axes
plt.xlabel(r'$ln$ $E$ (en $keV$) ')
plt.ylabel(r'$ln$ $R_{relatif}$')
axes = plt.gca()
axes.yaxis.set_tick_params(direction='in')
axes.xaxis.set_tick_params(direction='in')


# coefficient de la régression
plt.text(3.22, -4.33, r'$ln$ $R_{relatif} = ' + f'{float(modeleReg_1.coef_):.2f}$ $ln$ $E {float(modeleReg_1.intercept_):.2f}$')
plt.text(3.4, -4.4, r'$R^2 = ' +f'{modeleReg_1.score(pd.DataFrame(np.log(E_4)), np.log(Incertitude_4/E_4)):.3f}$')

# titre et affichage
plt.title(r'Evolution de $ln$ $R_{relatif} en fonction de $ln$ $E$')
plt.show()
```

VI) Spectre D'énergie

Extraction de donné d'un fichier

```python
def extraction(nom_fichier, nb_i, nb_f):
    with open(nom_fichier, 'r') as f:
        contenu = f.read()

    spectre = []

    # Extraire la première ligne du fichier
    for i in range(nb_i, nb_f):
        spectre.append(contenu.split('\n')[i])

    print(spectre)
    return spectre
```
